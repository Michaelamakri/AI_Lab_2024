## Τανυστές

Μία δομή δεδομένων που υλοποιεί η PyTorch και χρησιμοποιείται κατά κόρον όταν δουλεύουμε με αυτή ονομάζεται *τανυστής (tensor)*.

Μπορεί να φανταστεί κανείς έναν τανυστή σαν μία γενίκευση ενός διανύσματος ή πίνακα που, όμως, υποστηρίζει οποιονδήποτε αριθμό διαστάσεων. Με άλλα λόγια, όπως ένα διάνυσμα είναι μία μονοδιάστατη δομή δεδομένων και ένας πίνακας μία δισδιάστατη δομή δεδομένων, ένας τανυστής είναι μία *n*-διάστατη δομή δεδομένων, με το *n* να ανήκει στο σύνολο των φυσικών αριθμών.

> *Αυτός ο ορισμός δεν είναι σε καμία περίπτωση πλήρης και είναι χρήσιμος μόνο από μία προγραμματιστική σκοπιά. Η γραμμική άλγεβρα δίνει [πολλούς διαφορετικούς ορισμούς](https://en.wikipedia.org/wiki/Tensor#Definition) για τους τανυστές, η χρησιμότητα των οποίων εξαρτάται από το περιβάλλον στο οποίο τους χρησιμοποιούμε.*

Στην PyTorch, βέβαια, ένα αντικείμενο τύπου tensor είναι πολλά παραπάνω: ενώ μπορούμε να χειριστούμε τη δομή μέσω της python, εκτελώντας εκατοντάδες πράξεις, οι πράξεις αυτές εσωτερικά είναι υλοποιημένες σε C++, βελτιστοποιημένες ώστε να μπορούν να εκτελεστούν πάρα πολύ γρήγορα τόσο σε CPU όσο και σε GPU.

Για το λόγο αυτό, όταν τυπώσουμε τον τανυστή, μπορεί οι τιμές των στοιχείων να είναι 0 ή να είναι τυχαίοι αριθμοί (εξαρτάται από τα δεδομένα που έτυχε να βρίσκονται ήδη στη θέση μνήμης που δεσμεύσαμε).

Όμοια, μπορούμε να δημιουργήσουμε έναν δισδιάστατο τανυστή (δηλαδή έναν πίνακα) με 3 γραμμές και 4 στήλες:

````python
torch.tensor([[1, 2, 3], (4, 5, 6)])
````

Ίσως προσέξατε ότι, σε αντίθεση με πριν, οι αριθμοί δεν έχουν "." στο τέλος. Αυτός είναι ο τρόπος της python να μας δείξει ότι οι αριθμοί δεν είναι δεκαδικοί αλλά ακέραιοι.

Γίνεται, δηλαδή, αντιληπτό πως ένας τανυστής υποστηρίζει την αποθήκευση συγκεκριμένου τύπου δεδομένων. Αυτό τον τύπο δεδομένων προσπαθεί να τον εκμαιεύσει μόνη της η PyTorch.

Ο προεπιλεγμένος τύπος δεδομένων είναι ο `torch.float32`, δηλαδή δεκαδικός αριθμός μεγέθους 32 bit. Οι συνηθέστεροι τύποι δεδομένων είναι οι ακόλουθοι:
Τύπος δεδομένων | Περιγραφή
----------------|--------------------------------------------------
`torch.float16` | Δεκαδικός αριθμός μεγέθους 16 bit.
`torch.float32` | Δεκαδικός αριθμός μεγέθους 32 bit.
`torch.float64` | Δεκαδικός αριθμός μεγέθους 64 bit.
`torch.int8`    | Ακέραιος αριθμός μεγέθους 8 bit.
`torch.int16`   | Ακέραιος αριθμός μεγέθους 16 bit.
`torch.int32`   | Ακέραιος αριθμός μεγέθους 32 bit.
`torch.int64`   | Ακέραιος αριθμός μεγέθους 64 bit.
`torch.uint8`   | Ακέραιος μη-προσημασμένος αριθμός μεγέθους 8 bit.
`torch.bool`    | Λογική (Boolean) τιμή (True/False).

Μπορούμε να ορίσουμε ρητά τον τύπο δεδομένων ενός τανυστή κατά τη δημιουργία του ως εξής:

````python
torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
````

Εάν έχουμε ήδη έναν τανυστή με τα δεδομένα που θέλουμε αλλά με λάθος τύπο δεδομένων, μπορούμε να δημιουργήσουμε ένα νέο τανυστή με διαφορετικό τύπο δεδομένων χρησιμοποιώντας τη μέθοδο `to`.

````python
x = torch.tensor([[0, 1, 2], [3, 4, 5]])
print("x:", x.dtype)

y = x.to(torch.float32)
print("y:", y.dtype)
````

Ενδιαφέρον παρουσιάζει ο πολλαπλασιασμός τανυστών μεταξύ τους. Ας δημιουργήσουμε δύο τανυστές:

````python
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
y = torch.tensor([[1, 2], [3, 4], [5, 6]])

print(x, x.shape, end='\n\n')
print(y, y.shape, end='\n\n')
````

````python
x * x
````

Για να πραγματοποιήσουμε τον κλασικό πολλαπλασιασμό τανυστών/πινάκων, χρησιμοποιούμε το σύμβολο `@`.

````python
x @ y
````

Όταν εκτελούμε πράξεις μεταξύ τανυστών χρειάζεται μεγάλη προσοχή στο σχήμα των τανυστών ώστε να είναι τέτοιο που επιτρέπει να εκτελεστεί η εκάστοτε πράξη. Τα παρακάτω δύο κελιά προκαλούν `RuntimeError`.

````python
x * y
````

````python
x @ x
````

Υπάρχουν πολλές ακόμη συναρτήσεις που κάνουν δυνατή την εκτέλεση τριγωνομετρικών πράξεων, πράξεων μεταξύ bit (bitwise operations), στατιστικής ανάλυσης, πράξεων γραμμικής άλγεβρας κ.ά. Όλες αυτές μπορούν να βρεθούν στο [documentation της PyTorch](https://pytorch.org/docs/stable/torch.html#math-operations).

### Αντιγραφή τανυστών

Κάποιες φορές είναι επιθυμητό να αντιγράψουμε έναν τανυστή σε μία άλλη μεταβλητή.

````python
x = torch.zeros(8)
y = x

y[0] = 1

print('x:', x)
print('y:', y)
````

Ο τελεστής `=` δε δημιουργεί ένα νέο τανυστή. Κάνει τη μεταβλητή `y` να δείχνει στην ίδια θέση μνήμης με τη μεταβλητή `x`. Δηλαδή, τόσο η `x` όσο και η `y` αναφέρονται στον ίδιο τανυστή.

Εάν επιθυμούμε να έχουμε ένα πραγματικό αντίγραφο του τανυστή, το οποίο να μπορούμε να τροποποιούμε ανεξάρτητα, χρησιμοποιούμε τη μέθοδο `clone`.

````python
x = torch.zeros(8)
y = x.clone()

y[0] = 1

print('x:', x)
print('y:', y)
````

### Τανυστές στην GPU

Το μεγάλο πλεονέκτημα των τανυστών της PyTorch είναι ότι μπορούν να αποθηκευτούν στη μνήμη της GPU, ώστε στην GPU να εκτελεστούν οι επιθυμητές πράξεις, πετυχαίνοντας τεράστια επιτάχυνση για πολύ μεγάλους τανυστές.

Για να μπορούμε να εκμεταλλευτούμε αυτό το πλεονέκτημα, είναι απαραίτητο να έχουμε μία κάρτα γραφικών της Nvidia που υποστηρίζει την αρχιτεκτονική CUDA και να είναι εγκατεστημένοι οι ανάλογοι drivers. Για να ελέγξουμε ότι το σύστημά μας ικανοποιεί τα παραπάνω, μπορούμε να τρέξουμε την παρακάτω συνάρτηση:

````python
torch.cuda.is_available()
````

Χρησιμοποιώντας την παραπάνω συνάρτηση μπορούμε να ορίσουμε τη συσκευή στην οποία θα δουλέψουμε:

````python
if torch.cuda.is_available():
    my_device = torch.device('cuda')
else:
    my_device = torch.device('cpu')

print(my_device)
````

Κατά τη δημιουργία ενός τανυστή, μπορούμε να ορίσουμε τη συσκευή στην οποία θέλουμε να τον αποθηκεύσουμε:

````python
x = torch.tensor([1, 2, 3, 4], device=my_device)
print(x, x.device)
````

Έχουμε, ακόμη, τη δυνατότητα να δημιουργήσουμε ένα αντίγραφο ενός τανυστή που έχουμε ήδη δημιουργήσει σε μία άλλη συσκευή χρησιμοποιώντας τη μέθοδο `to`:

````python
x = torch.tensor([1, 2, 3, 4])
y = x.to(my_device)
print(y, y.device)
````

### Datasets

Ένα dataset είναι απλά μία συλλογή δεδομένων με τη μορφή κλάσης. Έχουν, όμως, τέτοια δομή ώστε να μας επιτρέπουν να διαχειριζόμαστε καθολικά τα δεδομένα μας, με μικρή εξάρτηση από την εκάστοτε εφαρμογή.

Μία κλάση τύπου dataset έχει τα ακόλουθα χαρακτηριστικά:
- Κληρονομεί την αρχέγονη (primitive) κλάση `torch.utils.data.Dataset`.
- Υλοποιεί τη μέθοδο `__init__` που φορτώνει τα δεδομένα στη μνήμη.
- Υλοποιεί τη μέθοδο `__len__` που επιστρέφει το μέγεθος του dataset.
- Υλοποιεί τη μέθοδο `__getitem__` που επιστρέφει ένα δείγμα (sample) από το dataset.

Για να κατανοήσουμε τα παραπάνω, ας υλοποιήσουμε μία κλάση που αναπαριστά ένα dataset το οποίο φορτώνει από ένα αρχείο ζεύγη αριθμών x,y που προέρχονται από μία μυστηριώδη συνάρτηση.

````python
# Θα συμπληρωθεί την ώρα του εργαστηρίου

class MysteryFunctionDataset(Dataset):
    def __init__(self, file_path):

        self.x = []
        self.y = []

        with open(file_path) as f:
            for line in f:
                x, y = line.split(',')
                x, y = float(x), float(y)

                self.x.append(x)
                self.y.append(y)

    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, index):
        return self.x[index], self.y[index]
````

Τώρα, μπορούμε να δημιουργήσουμε ένα καινούργιο στιγμιότυπο της κλάσης:

````python
dataset = MysteryFunctionDataset('mystery.txt')
````

Μπορούμε να ανακτήσουμε δείγματα των δεδομένων μας από αυτό:

````python
dataset[0]
````

Ή να το διατρέξουμε μέσα σε ένα βρόχο επανάληψης:

````python
for x, y in dataset:
    print(x, y)
````

### DataLoaders

Οι DataLoaders είναι δομές που διαχειρίζονται ένα dataset παρέχοντάς μας δυνατότητες που είναι συχνά επιθυμητές όταν δουλεύουμε με νευρωνικά δίκτυα, όπως τη λήψη των δειγμάτων σε παρτίδες (*minibatches*) και το αυτόματο ανακάτεμα των δειγμάτων κάθε φορά που ολοκληρώνουμε τη λήψη όλων των δειγμάτων του dataset.

Για παράδειγμα, ας δημιουργήσουμε έναν DataLoader για το dataset που φτιάξαμε που να επιστρέφει 100 δείγματα κάθε φορά και να τα ανακατεύει σε κάθε επανάληψη του dataset.

````python
loader = DataLoader(dataset, batch_size=100, shuffle=True)
````

Με τον παρακάτω τρόπο μπορούμε να πάρουμε το επόμενο κάθε φορά batch από δείγματα του dataset:

````python
next(iter(loader))
````
